```python
from collections import Counter

# Go through each line in the text and count the number of times each LINE appears
counter = Counter(text.split("\n"))

counter.most_common(10)
```

```python
# Get the most common lines from the counters
most_common_lines = [line for line, _ in counter.most_common(8)]

# Split the text into lines
lines = text.split("\n")

# Remove the lines that match the most common lines
filtered_lines = [line for line in lines if line not in most_common_lines]

# Join the filtered lines back into a single text
filtered_text = "\n".join(filtered_lines)

# save new text to file
with open("filtered_text.txt", "w", encoding='utf-8') as file:
    file.write(filtered_text)
```

```python
# Go through each line in the text and count the number of times each LINE appears
counter = Counter(filtered_text.split("\n"))

counter.most_common(20)
```

```python
import re

# Define the pattern for dates like so: 31/10/17   5:39 PM
date_pattern = r"\b\d{2}/\d{2}/\d{2}\s+\d{1,2}:\d{2}\s(?:AM|PM)\b"

for line in filtered_text.split("\n"):
    if re.search(date_pattern, line):
        # Remove lines that contain the date pattern
        filtered_text = filtered_text.replace(line, "")

# Remove any empty lines
filtered_text = "\n".join([line for line in filtered_text.split("\n") if line])

lines_to_remove = ['www.cengagebrain.com. ', 'www.cengagebrain.com.']

filtered_text = "\n".join([line for line in filtered_text.split("\n") if line not in lines_to_remove])

# Go through each line in the text and count the number of times each LINE appears
counter = Counter(filtered_text.split("\n"))

counter.most_common(20)
```
